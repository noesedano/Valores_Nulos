```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as pit
```

```python
#carga desde un archivo csv
data =pd.read_csv("Ventas_totales.csv")
data.head(5)
```

```python
#identificar  valores nulos por columna
valores_nulos=data.isnull().sum()
valores_nulos
```

REALIZAMOS  UNA COPIA DEL DATAFRAME

```python
#Realizamos la copia del dataframe
data3=data.copy()
```

```python
#primer metodo de sustitucion de valores nulos
#sustituir valores nujlos con promedio o media
data3["salon_ventas"]=data3["salon_ventas"].fillna(round(data["salon_ventas"].mean(),1))
```

```python
#identificar  valores nulos por dataframe
valores_nulos=data3.isnull().sum()
valores_nulos
```

```python
#convertir dataframe a csv
data3.to_csv("nulo_promedio.csv")
```

```python
#segundo metodo de sustitucion de valores nulos
#sustituir valores nulos con mediana
data3["tarjetas_debito"]=data3["tarjetas_debito"].fillna(round(data["tarjetas_debito"].median(),1))
```

```python
#identificar  valores nulos por dataframe
valores_nulos=data3.isnull().sum()
valores_nulos
```

```python
#convertir dataframe a csv
data3.to_csv("nulo_media_tarjeta_debito.csv")
```

```python
#tercer metodo de sustitucion de valores nulos
#sustituir valores nulos por un valor numerico en concreto
data3["tarjetas_credito"]=data3["tarjetas_credito"].fillna(10)
data3
```

```python
#identificar  valores nulos por dataframe
valores_nulos=data3.isnull().sum()
valores_nulos
```

```python
#convertir dataframe a csv
data3.to_csv("nulo_numerico_tarjetas_credito.csv")
```

```python
#cuarto  metodo de sustitucion de valores nulos
#sustituir valores nulos por un valor string en concreto
data3["otros_medios"]=data3["otros_medios"].fillna("2030-06-19 00:00:00")
```

```python
#identificar  valores nulos por dataframe
valores_nulos=data3.isnull().sum()
valores_nulos
```

```python
#quinto  metodo de sustitucion de valores nulos
#sustituir valores nulos por valores no nulos hacia adelante "forward fill" 
#filtro por columnas
data3["subtotal_ventas_alimentos_bebidas"]=data3["subtotal_ventas_alimentos_bebidas"].fillna(method="ffill")
```

```python
#identificar  valores nulos por dataframe
valores_nulos=data3.isnull().sum()
valores_nulos
```
